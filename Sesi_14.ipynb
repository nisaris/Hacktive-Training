## PCA

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
bc_data = data.data
```

```python
bc_data.shape
```

```python
bc_labels = data.target
```

```python
bc_labels.shape
```

```python
import numpy as np
import pandas as pd
```

```python
labels = np.reshape(bc_labels, (569, 1))
```

```python
final_data = np.concatenate([bc_data, labels], axis=1)
```

```python
final_data.shape
```

```python
bc_dataset = pd.DataFrame(final_data)
```

```python
bc_dataset.head()
```

```python
features = data.feature_names
```

```python
features_labels = np.append(features, 'label')
```

```python
features_labels
```

```python
bc_dataset.columns = features_labels
```

```python
bc_dataset.head()
```

```python
bc_dataset['label'].replace(0, 'Benign', inplace=True)
bc_dataset['label'].replace(1, 'Malignant', inplace=True)
```

```python
bc_dataset.tail()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = bc_dataset.loc[:, features].values
```

```python
x = StandardScaler().fit_transform(x)
```

```python
normalized_data = pd.DataFrame(x)
normalized_data.tail()
```

```python
from sklearn.decomposition import PCA
pca_bc = PCA(n_components=2)
principalComponents_bc = pca_bc.fit_transform(x)
```

```python
principalComponents_bc
```

```python
principal_DF = pd.DataFrame(data=principalComponents_bc,
                            columns=['principal component 1', 'principal component 2'])
```

```python
principal_DF.tail()
```

```python
pca_bc.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.figure(figsize=(10, 10))
plt.xlabel('PC1')
plt.ylabel('PC2')

targets = ['Benign', 'Malignant']
colors = ['r', 'g']

for target, color in zip(targets, colors):
    warna = bc_dataset['label'] == target
    plt.scatter(principal_DF.loc[warna, 'principal component 1'],
           principal_DF.loc[warna, 'principal component 2'],
           c = color, s = 50)
    
plt.legend(targets)

plt.show()
```

## K-Means

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

%matplotlib inline
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv')
data.head()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler

data_scaled = StandardScaler().fit_transform(data)
```

```python
pd.DataFrame(data_scaled).describe()
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(data_scaled)
```

```python
kmeans.inertia_
```

```python
elbow = []

for cluster in range(1, 20):
    kmeans = KMeans(n_clusters=cluster)
    kmeans.fit(data_scaled)
    elbow.append(kmeans.inertia_)
```

```python
frame = pd.DataFrame({'Cluster': range(1, 20), 'SSE':elbow})
frame
```

```python
plt.figure(figsize=(12, 6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of Cluster')
plt.ylabel('Inertia')
```

```python
kmeans = KMeans(n_clusters=6)
kmeans.fit(data_scaled)
kmeans.inertia_
```

```python
pred = kmeans.predict(data_scaled)
```

```python
frame = pd.DataFrame(data_scaled)
frame['cluster'] = pred
frame.head()
```

```python
frame['cluster'].value_counts()
```
