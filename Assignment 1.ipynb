```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['ggplot'])
import datetime as datetime

%matplotlib inline
```

```python
fb_pelatihan = pd.read_excel('dataset/Feedback_Lat_2020.xlsx')
```

```python
fb_pelatihan.head(5)
```

```python
fb_pelatihan.info()
```

```python
fb_pelatihan['Academy Event'] = fb_pelatihan['Academy Event'].astype(str)
fb_pelatihan['Tgl Mulai'] = pd.to_datetime(fb_pelatihan['Tgl Mulai'])
fb_pelatihan['Bulan'] = fb_pelatihan['Tgl Mulai'].dt.month
fb_pelatihan.info()
```

# Area Plot

Sekarang kita akan membuat area plot dari jumlah peserta academy DSP dan Consumer per bulan untuk melihat di waktu kapan, kedua akademi tersebut melayani peserta terbanyak

```python
d_area = fb_pelatihan[['JML Peserta', 'Academy Event', 'Bulan']]
area_dsp = d_area[d_area['Academy Event'].isin(['DSP'])]
area_dsp.rename(columns = {'JML Peserta' : 'Jumlah Peserta DSP'}, inplace = True)

area_cons = d_area[d_area['Academy Event'].isin(['CONSUMER'])]
area_cons.rename(columns = {'JML Peserta' : 'Jumlah Peserta Consumer'}, inplace = True)
```

```python
area_dsp.head(3)
```

```python
area_cons.head(3)
```

```python
data_1 = area_dsp.groupby('Bulan', axis = 0)['Jumlah Peserta DSP'].sum()
data_2 = area_cons.groupby('Bulan', axis = 0) ['Jumlah Peserta Consumer'].sum()
```

```python
print(data_1)
print(data_2)
```

```python
data_area = pd.concat([data_1, data_2], axis = 1, sort = False)
data_area
```

```python
data_area.plot(kind = 'area',
              stacked = False,
              figsize = (13,7))

plt.title('Jumlah Peserta Akademi DSP dan Consumer tahun 2020')
plt.ylabel('Jumlah Peserta')
plt.xlabel('Bulan')

plt.show()
```

Dapat dilihat dari area plot tersebut, jika Akademi consumer cenderung melayani peserta lebih banyak dari pada akademi DSP dan mencapai puncaknya pada bulan Agustus, sedangkan akademi DSP mencapai puncaknya pada bulan Juni. Dari situ dapat dilihat bila pelatihan cenderung lebih banyak di pertengahan tahun.

# Histogram

Kita akan melihat sebaran pengisian NPS pada akademi DSP

```python
hist_dsp = fb_pelatihan[fb_pelatihan['Academy Event'].isin(['DSP'])]
hist_dsp.head(3)
```

```python
count, bin_edges = np.histogram(hist_dsp['Net Promotor Score'], 10)
hst = hist_dsp['Net Promotor Score'].plot(kind = 'hist',
                                         figsize = (15,10),
                                         bins = 10,
                                         xticks = bin_edges,
                                          color = ['skyblue'])

hst.set_title('Net Promotor Score Akademi DSP')
hst.set_xlabel('Net Promotor Score')
plt.show()
```

Dari gambar tersebut dapat dilihat apabila di akademi DSP nilai NPS 100 sudah cukup banyak, walaupun di sisi lain masih banyak pula yang memberikan NPS dengan nilai 0 - 20 dan perlu menjadi perhatian lebih lanjut.
Disamping itu ada hal menarik, karena ada anomali data, dimana ada nilai NPS yang dibawa 0, padahal seharusnya nilai NPS adalah di range 0 - 100. Ini dapat mengindikasikan adanya error pada sistem cognitium serta sistem input nilai UBPP/NPS.

# Pie Chart

Kita akan membuat pie chart terkait proporsi event di tahun 2020, dimana ada event LAT atau berbentuk pelatihan, dan ada event NON LAT yang merupakan event selain event pelatihan.

```python
cat_pelat = fb_pelatihan.groupby('Category', axis = 0)['Nama Pelatihan'].count()
```

```python
pie_event = cat_pelat.plot(kind = 'pie',
                          figsize = (20,10),
                          autopct = '%1.1F%%',
                          startangle = 90,
                          labels = None,
                          pctdistance = 1.12,
                          colors = ['lightgreen', 'purple'],
                          explode = [0, 0.1]
                          )

pie_event.set_title('Jumlah Event Berdasarkan Jenis Event')
pie_event.axis('equal')
pie_event.set_ylabel('Event Pelatihan')
pie_event.legend(labels = cat_pelat.index, loc = 'upper right')
plt.show()
```

Dari gambar tersebut kita dapat melihat jika proporsi event pelatihan lebih banyak dibanding event non pelatihan. Walaupun begitu, jumlah event non pelatihan juga cukup banyak yaitu 27,1%.

# Bar Chart

Kita akan melihat jumlah peserta dari setiap akademi dengan menggunakan bar chart.

```python
jml_peserta = fb_pelatihan.groupby('Academy Event', sort = False)['JML Peserta'].sum()
jml_peserta
```

```python
bar = jml_peserta.plot(y = 'JML Peserta', x = 'Academy Event',
                      kind = 'barh',
                      alpha = 0.7,
                      color = 'red',
                      figsize = (15,10))
bar.set_title('Jumlah Peserta per Unit 2020')
bar.set_ylabel('Nama Unit')
bar.set_xlabel('Jumlah Peserta')
plt.show()
```

Dari gambar tersebut dapat dilihat jika jumlah peserta terbanyak adalah dari akademi Business Enabler dan yang paling sedikit adalah dari Mobile / kmcs / GoFM, dImana memang KMCS dan GoFM merupakan unit non akademi sehingga event yang diadakan memang lebih sedikit.

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
nama_pelatihan = open('dataset/Judul_ELearning.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
namapelatihan_wc = WordCloud(
    background_color='white',
    max_words=200,
    stopwords=stopwords
)

namapelatihan_wc.generate(nama_pelatihan)
```

```python
stopwords.add('Batch')
stopwords.add('Certified')
stopwords.add('Sesi')


namapelatihan_wc.generate(nama_pelatihan)

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(namapelatihan_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Dilihat dari wordcloud tersebut, dapat dilihat jika e-learning yang sering diakses adalah wawasan kebangsaan, CX transformation, transformation program, communication skill, dan juga telkom codes.
