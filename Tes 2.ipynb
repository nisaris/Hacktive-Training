```python
fb_pelatihan = pd.read_excel('./dataset/Feedback_Lat_2020.xlsx', sheet_name='UBPP Penyelenggaraan', skiprows=range(0), skipfooter=2)
```

```python
fb_pelatihan.head()
```

```python
fb_UBPP=fb_pelatihan[['Nama Pelatihan', 'Academy Event', 'Tgl Mulai', 'UBPP Inst', 'UBPP Akom', 'UBPP Sarana', 'UBPP Laborat', 'UBPP Penyelenggaraan', 'UBPP Kafetaria', 'UBPP Materi']]
```

```python
fb_UBPP.head()
fb_pelatihan['Tgl']=fb_pelatihan['Tgl Mulai'].astype('datetime64[ns]')
fb_pelatihan.head()
```

```python
AVG_UBPP=fb_UBPP.groupby('Academy Event').mean()
AVG_UBPP.head()
```

Area Plot

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

print ('Matplotlib version: ', mpl.__version__)
```

```python
print(plt.style.available)
```

```python
mpl.style.use(['ggplot'])
```

```python
AVG_UBPP=AVG_UBPP.transpose()
```

```python
AVG_UBPP.plot(kind='area',
              alpha = 0.45,
              stacked = True,
              figsize = (20,10),)

plt.title('Rata-Rata UBPP per Academy')
plt.ylabel('Rata-Rata')
plt.xlabel('UBPP')

plt.show()
```
